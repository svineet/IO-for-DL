```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt

%matplotlib inline
```

```python
cmds = pd.read_csv("unique_commands.csv")
```

```python
cmds
```

```python
cmds['Area'] = "Not Available"
cmds['Science'] = "Not Available"

cmds
```

```python
proj = pd.read_csv("TitanRUR/ProjectAreas.csv")
proj
```

```python
for i, row in proj.iterrows():
    area_code = row['ID']
    filt = cmds['command'].str.contains(area_code)
    
    cmds["Area"][filt] = row['Area']
    cmds["Science"][filt] = row['Science']
```

```python
cmds
```

```python
cmds['command'] = cmds['command'].apply(lambda s: s if "/" not in s else s.split("/")[-1])
```

```python
cmds
```

```python
cmds.to_csv("labelled_commands.csv", index=False)
```
